```python
import numpy as np
import pandas as pd
import os
from yaml import safe_load
from tqdm import tqdm
import pickle
```

```python
filenames = []
for file in os.listdir('odi_data'):
    filenames.append(os.path.join('odi_data', file))
```

```python
filenames[0:10]
```

```python
final_df = pd.DataFrame()
counter = 1
for file in tqdm(filenames):
    with open(file, 'r') as f:
        df = pd.json_normalize(safe_load(f))
        df['match_id'] = counter
        final_df = pd.concat([final_df, df], ignore_index=True)
        counter += 1
        
```

```python
match = pickle.load(open('dataset_1v.pkl', 'rb'))
```

```python
match.head()
```

```python
match.iloc[0]['innings'][0]['1st innings']['deliveries']
```

```python
import pandas as pd

count = 1
delivery_df = pd.DataFrame()

for index, row in match.iterrows():
    if count in [75, 108, 150, 180, 268, 360, 443, 458, 584, 748, 982, 1052, 1111, 1226, 1345]:
        count += 1
        continue
    count += 1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []

    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')

    loop_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue
    })

    delivery_df = pd.concat([delivery_df, loop_df], ignore_index=True)
```

```python
delivery_df
```

```python
delivery_df.info()
```

```python
def feilding_team(rows):
    for team in rows['teams']:
        if team != rows['batting_team']:
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(feilding_team, axis=1)
```

```python
delivery_df.drop(columns=['teams'],inplace=True)
```
